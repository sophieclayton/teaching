This practical is based on Riley's 1946 model of phytoplankton population dynamics on the George's Bank:  
Riley, G. A. (1946) Factors Controlling Phytoplankton Populationson George's Bank. *J. Mar. Res.* **6**, pp 54-73


```python
# These few lines of code import all of the different packages that are needed to run code and make nice plots
# You don't need to change anything here!
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Mean data for top 30m - DON"T CHANGE ANY OF THIS!

YD = [15,75,105,135,165, 255] # yearday, assuming a simplified calendar with 12 30-day months
T = [4.61, 2.60, 3.81, 5.14, 9.66, 15.24] # temperature
P = [33.7, 34.7, 21.9, 16.6, 19.2, 14.4] # phosphate
N = [209, 172, 129, 285, 155, 153] # nitrate
Z = [14, 24, 32, 106, 103, 135] # zooplankton, thousands per m^2
Ph = [118, 828, 2303, 871, 478, 560] # phytoplankton pigments, thousands of Harvey units per m^2
```

```python
# Let's make a plot of Riley's data

fig1, ax1 = plt.subplots()
ax1.plot(YD, T, 'r.-')
ax1.set_xlabel('Yearday')
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Temperature (C)', color='r')
for tl in ax1.get_yticklabels():
    tl.set_color('r')


ax2 = ax1.twinx()
ax2.plot(YD, Ph, 'g.-')
ax2.set_ylabel('Phytoplankton', color='g')
for tl in ax2.get_yticklabels():
    tl.set_color('g')
plt.show(fig1)

## Now try changing the plot to show different variables, e.g. phytoplankton and zooplankton, 
## and then run this cell again to see the plot
```
